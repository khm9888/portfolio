## 승차/하차 테이블 분리

### 승차 테이블 만들기

```python
df
```

```python
# 공통 컬럼
df1 = df.iloc[:,:3]
df1
```

```python
# 승차 컬럼만 가져오기
df2 = df.iloc[:,3::2]
df2
```

```python
df2.columns.str.split(" ")
```

```python
df2.columns = df2.columns.str.split(" ").str[0]
```

```python
# 공통컬럼과 승차컬럼 연결하기
df_in = pd.concat([df1,df2],axis=1)
df_in
```

## 하차 테이블 만들기

```python
# 공통 컬럼
df1
# 하차 컬럼만 가져오기
df3 = df.iloc[:,4::2]
```

```python
df3.columns = df3.columns.str.split(" ").str[0]
```

```python
# 공통컬럼과 하차컬럼 연결하기
df_out = pd.concat([df1,df3],axis=1)
df_out
```


# 강남역의 최근 시간대별 승하차정보 분석

## 강남역의 최근 승차정보 분석

```python
# 강남역의 최근 승차 데이터 불러오기 
df_in_202108_gangnam = df_in_202108[df_in_202108["지하철역"]=="강남"].iloc[:,3:]
df_in_202108_gangnam
```

```python
# melt
df_in_202108_gangnam=df_in_202108_gangnam.melt()
```

```python
# 컬럼명 변경
df_in_202108_gangnam.columns = ["시간대","이용수"]
df_in_202108_gangnam.sort_values("이용수",ascending=False)
```

```python
# 시간대별 승차인원 시각화하기
x = df_in_202108_gangnam["시간대"]
y = df_in_202108_gangnam["이용수"]
plt.bar(x,y)
plt.title("강남역 이용수 분석")
plt.xticks(rotation=45)
plt.show()
```

```python
# 시간대별 승차인원 시각화하기
x = df_in_202108_gangnam["시간대"]
y = df_in_202108_gangnam["이용수"]
plt.barh(x,y)
plt.title("강남역 이용수 분석")
plt.xticks(rotation=45)
plt.show()
```
